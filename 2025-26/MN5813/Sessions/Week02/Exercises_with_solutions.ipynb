### Exercise 4: Tuple operations

Create a tuple `colors` containing five different color names. Then:

1. Print the entire tuple
2. Access and print the third color
3. Attempt to change the second color (this should raise an error)
4. Create a new tuple by concatenating `colors` with another tuple containing two more colors
5. Count how many times 'blue' appears in the tuple (case-insensitive)

Hint: Remember that tuples are immutable. Use tuple concatenation with the + operator and the `count()` method.

```python
# Solution
colors = ("red", "blue", "green", "yellow", "purple")

# 1. Print the entire tuple
print(f"1. Entire tuple: {colors}")

# 2. Access and print the third color
print(f"2. Third color: {colors[2]}")

# 3. Attempt to change the second color (this will raise an error)
try:
    colors[1] = "orange"
except TypeError as e:
    print(f"3. Error when trying to change tuple: {e}")

# 4. Create a new tuple by concatenating colors with another tuple
more_colors = ("pink", "brown")
all_colors = colors + more_colors
print(f"4. Concatenated tuple: {all_colors}")

# 5. Count how many times 'blue' appears in the tuple (case-insensitive)
blue_count = sum(1 for color in all_colors if color.lower() == "blue")
print(f"5. 'Blue' appears {blue_count} time(s) in the tuple")
```

## Making decisions in your programmes

### Exercise 5: Grade calculator

Write a program that takes a numerical grade as input and returns the corresponding letter grade according to the following scale:

- A: 90-100
- B: 80-89
- C: 70-79
- D: 60-69
- F: below 60

Hint: Use if-elif-else statements to check the range of the input grade.

```python
def calculate_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Test the function
test_scores = [110, 82, 75, 65, 55]
for score in test_scores:
    print(f"Score: {score}, Grade: {calculate_grade(score)}")
```

### Exercise 6: Leap year checker

Write a function that takes a year as input and returns True if it's a leap year, and False otherwise. A leap year is divisible by 4, except for century years, which must be divisible by 400.

Hint: Use the modulo operator (%) to check divisibility. Remember to handle century years correctly.

```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

# Test the function
test_years = [2000, 2020, 2100, 2200, 2300, 2400]
for year in test_years:
    print(f"{year} is leap year: {is_leap_year(year)}")
```

### Exercise 8: Fibonacci sequence

Write a function that generates the first n numbers in the Fibonacci sequence. The Fibonacci sequence starts with 0 and 1, and each subsequent number is the sum of the two preceding ones.

Hint: Use a while loop to generate the sequence. Keep track of the two previous numbers and calculate the next one in each iteration.

```python
def fibonacci(n):
    sequence = []
    a, b = 0, 1
    while len(sequence) < n:
        sequence.append(a)
        a, b = b, a + b
    return sequence

# Test the function
print(fibonacci(10))
```

### Exercise 9: Temperature converter

Write two functions:

1. `celsius_to_fahrenheit(celsius)` which converts Celsius to Fahrenheit
2. `fahrenheit_to_celsius(fahrenheit)` which converts Fahrenheit to Celsius

Then write a main function that asks the user for a temperature and a conversion direction (C to F or F to C), calls the appropriate function, and prints the result.

Hint: The formulas are:
- F = (C * 9/5) + 32
- C = (F - 32) * 5/9

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def main():
    temp = float(input("Enter the temperature: "))
    direction = input("Enter 'C' for Celsius to Fahrenheit, or 'F' for Fahrenheit to Celsius: ").upper()
    
    if direction == 'C':
        result = celsius_to_fahrenheit(temp)
        print(f"{temp}°C is equal to {result:.2f}°F")
    elif direction == 'F':
        result = fahrenheit_to_celsius(temp)
        print(f"{temp}°F is equal to {result:.2f}°C")
    else:
        print("Invalid input. Please enter 'C' or 'F'.")

# Uncomment the line below to run the main function
# main()
```

### Exercise 10: Calculator function

Write a function called `calculator` that takes two numbers and an operator (+, -, *, /) as arguments and returns the result of the operation. Include error handling for division by zero and invalid operators.

Hint: Use a dictionary to map operators to their corresponding operations, or use if-elif statements. Use a try-except block to handle the ZeroDivisionError.

```python
def calculator(num1, num2, operator):
    operations = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '*': lambda x, y: x * y,
        '/': lambda x, y: x / y
    }
    
    if operator not in operations:
        return "Error: Invalid operator"
    
    try:
        return operations[operator](num1, num2)
    except ZeroDivisionError:
        return "Error: Division by zero"

# Test the function
print(calculator(10, 5, '+'))
print(calculator(10, 5, '-'))
print(calculator(10, 5, '*'))
print(calculator(10, 5, '/'))
print(calculator(10, 0, '/'))
print(calculator(10, 5, '%'))
```

### Exercise 11: List Operations

Create a list of 10 random integers between 1 and 100. Then perform the following operations:

1. Print the original list
2. Sort the list in ascending order
3. Reverse the sorted list
4. Add the number 50 to the end of the list
5. Remove any number that is a multiple of 3
6. Print the final list

Hint: Use the `random` module to generate random numbers. Use list methods like `sort()`, `reverse()`, `append()`, and list comprehension for filtering.

```python
import random

# Create a list of 10 random integers between 1 and 100
numbers = [random.randint(1, 100) for _ in range(10)]

# 1. Print the original list
print("Original list:", numbers)

# 2. Sort the list in ascending order
numbers.sort()
print("Sorted list:", numbers)

# 3. Reverse the sorted list
numbers.reverse()
print("Reversed list:", numbers)

# 4. Add the number 50 to the end of the list
numbers.append(50)
print("List after adding 50:", numbers)

# 5. Remove any number that is a multiple of 3
numbers = [num for num in numbers if num % 3 != 0]

# 6. Print the final list
print("Final list:", numbers)
```

### Exercise 13: Even and odd numbers

Create a new list 'numbers' with 10 integer numbers. Find all the even and odd numbers and put them in two separate lists: `even_list` which will have all even numbers and `odd_list` that will have all odd numbers of the list 'numbers'.

Hint: Use the modulo operator (%) to check if a number is even or odd. If number % 2 == 0, it's even.

```python
# Create a list of 10 integer numbers
numbers = [3, 8, 12, 7, 15, 2, 20, 9, 14, 5]

# Separate even and odd numbers
even_list = [num for num in numbers if num % 2 == 0]
odd_list = [num for num in numbers if num % 2 != 0]

print("Original list:", numbers)
print("Even numbers:", even_list)
print("Odd numbers:", odd_list)
```

### Exercise 14: Dictionary of students

Create a dictionary to store information about students. Each student should have a name, age, and a list of subjects they're studying. Implement the following operations:

1. Add three students to the dictionary
2. Print all student names
3. Add a new subject for one of the students
4. Remove a subject from another student
5. Print the complete information for each student

Hint: Use nested dictionaries and lists. Use dictionary methods like `keys()` to get student names, and list methods to modify subjects.

```python
# Create a dictionary of students
students = {
    "Alice": {"age": 20, "subjects": ["Math", "Physics", "Computer Science"]},
    "Bob": {"age": 21, "subjects": ["Biology", "Chemistry", "Psychology"]},
    "Charlie": {"age": 19, "subjects": ["English", "History", "Art"]}
}

# 2. Print all student names
print("Student names:", list(students.keys()))

# 3. Add a new subject for Alice
students["Alice"]["subjects"].append("Statistics")

# 4. Remove a subject from Bob
students["Bob"]["subjects"].remove("Psychology")

# 5. Print the complete information for each student
for name, info in students.items():
    print(f"\nName: {name}")
    print(f"Age: {info['age']}")
    print(f"Subjects: {', '.join(info['subjects'])}")
```
