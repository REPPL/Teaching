## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set display options
pd.set_option("display.max_columns", 50)
plt.style.use("ggplot")
```

## Creating a data frame

We start by creating a sample DataFrame to work with throughout the demonstration.

```python
# Sample data
data = {
    "name": ["Alice", "Bob", "Carol", "Dan", "Eve"],
    "age": [25, 30, 35, 40, 45],
    "city": ["New York", "London", "Paris", "Tokyo", "Sydney"],
    "salary": [50000, 60000, 70000, 80000, 90000]
}

df = pd.DataFrame(data)
```

```python
print("Original DataFrame:")
print(df)
```

## Advanced indexing and selection

Here we demonstrate more complex data selection using boolean indexing and column selection.

### Boolean indexing

Here we demonstrate more complex data selection using boolean indexing and column selection. Boolean indexing allows you to select data based on conditions, creating a mask of True/False values that is then used to filter the DataFrame.

```python
print("Advanced indexing and selection:")
print(df.loc[df["age"] > 30, ["name", "city"]])
```

### Multi-level indexing

Multi-level (or hierarchical) indexing allows you to have multiple index levels on an axis. This is powerful for representing higher-dimensional data in a DataFrame.

```python
df_multi = df.set_index(["city", "name"])
```

```python
print("Multi-level indexing:")
print(df_multi)
```

```python
print("\nSelecting data for London:")
print(df_multi.loc["London"])
```

## Pivot tables

Pivot tables are used to summarize and aggregate data inside DataFrames. They're especially useful for data with multiple dimensions.

```python
sales_data = pd.DataFrame({
    "date": pd.date_range(start="2023-01-01", periods=10),
    "product": ["A", "B"] * 5,
    "sales": np.random.randint(100, 1000, 10)
})
pivot_table = pd.pivot_table(sales_data, values="sales", index="date", columns="product", aggfunc="sum")
print("Pivot table:")
print(pivot_table)
```

## Window functions and rolling calculations

Window functions allow you to perform calculations across a set of rows that are related to the current row. Rolling calculations are a type of window function that operate over a sliding window of data.

```python
df["salary_rolling_mean"] = df["salary"].rolling(window=2).mean()
print("DataFrame with rolling mean salary:")
print(df)
```

## Advanced data cleaning

This section demonstrates some advanced data cleaning techniques, including string manipulation and binning continuous data into categories.

```python
df["name_upper"] = df["name"].str.upper()
df["salary_bin"] = pd.cut(df["salary"], bins=[0, 60000, 80000, 100000], labels=["Low", "Medium", "High"])
print("DataFrame with advanced data cleaning:")
print(df)
```

## Working with MultiIndex

This section shows more advanced operations with MultiIndex DataFrames, including sorting and cross-sectional selection.

```python
df_multi.sort_index(inplace=True)
print("Sorted MultiIndex DataFrame:")
print(df_multi)
print("\nCross-section of MultiIndex DataFrame:")
print(df_multi.xs("London", level="city"))
```

## Advanced DataFrame operations

The `melt` function is used to reshape data from wide to long format. This is useful for converting columns into rows, which can be necessary for certain types of analysis or visualization.

```python
df_melted = pd.melt(df, id_vars=["name", "age"], value_vars=["salary", "salary_rolling_mean"])
print("Melted DataFrame:")
print(df_melted)
```

## Time series resampling

Resampling allows you to change the frequency of time series data. This is useful for aggregating high-frequency data to a lower frequency, or for converting between different time frequencies.

```python
time_series = pd.Series(np.random.randn(1000), index=pd.date_range(start="2023-01-01", periods=1000))
resampled = time_series.resample("W").mean()
print("Resampled time series (weekly mean):")
print(resampled.head())
```

## Advanced visualisation

This section demonstrates how to create more complex visualisations using `Pandas` and `Matplotlib`. Here, we're creating a scatter plot with points colored based on a categorical variable.

```python
plt.figure(figsize=(10, 6))
df.plot(x="age", y="salary", kind="scatter", c="salary_bin", cmap="viridis")
plt.title("Age vs Salary")
plt.colorbar(label="Salary Category")
plt.show()
```
